```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_data = open("Kalman_data_test_bla.txt","r")
```

```python
time = []
data = []

Acc_x = []
Acc_y = []
Acc_z = []

Gyro_x = []
Gyro_y = []
Gyro_z = []

Mag_x = []
Mag_y = []
Mag_z = []

sAx = 0.0
sAy = 0.0 
sAz = 0.0
sWx = 0.0
sWy = 0.0
sWz = 0.0

currentSample = 0
dt = 0.00635
```

```python
for l in file_data.readlines():
    try:
        temp = l.split(' ')

        Gyro_x.append(int(temp[0]))
        Gyro_y.append(int(temp[1]))
        Gyro_z.append(int(temp[2]))
    
        Acc_x.append(int(temp[3]))
        Acc_y.append(int(temp[4]))
        Acc_z.append(int(temp[5]))

        Mag_x.append(float(temp[6]))
        Mag_y.append(float(temp[7]))
        Mag_z.append(float(temp[8]))
            # data = np.array(data)
        currentSample += dt
        time.append(currentSample)
    except Exception as e:
        print(e)


file_data.close()
```

```python
Acc_x = np.array(Acc_x)
Acc_y = np.array(Acc_y)
Acc_z = np.array(Acc_z)

Gyro_x = np.array(Gyro_x)
Gyro_y = np.array(Gyro_y)
Gyro_z = np.array(Gyro_z)

Mag_x = np.array(Mag_x)
Mag_y = np.array(Mag_y)
Mag_z = np.array(Mag_z)

print(Acc_x.mean())
print(Acc_x.std())

print(Acc_y.mean())
print(Acc_y.std())

print(Acc_z.mean())
print(Acc_z.std())

print(Gyro_x.mean())
print(Gyro_x.std())

print(Gyro_y.mean())
print(Gyro_y.std())

print(Gyro_z.mean())
print(Gyro_z.std())
```

```python
plt.subplot(1, 1, 1)

plt.plot(time[0:100000], Gyro_x - Gyro_x.mean(), 'o-')
plt.title('Distribution of calibrated Gyro_x')
plt.ylabel('Gyro_x')

plt.show()

print("Gyro_x offset is : %0.2f",int(Gyro_x.mean()))
print("Standard deviation of Gyro_x (deg/s) is : %0.2f", (Gyro_x - Gyro_x.mean()).std()/16.4) 
sWx = (Gyro_x - Gyro_x.mean()).std()/16.4
print("Variance of Gyro_x is : %0.2f", np.square((Gyro_x - Gyro_x.mean()).std()/16.4))
```

```python
#Gyro drift effect on x:
gyro_angle = []
gyro_tmp = 0
for (i,t) in enumerate(time[0:100000]):
    gyro_tmp = gyro_tmp + dt*((Gyro_x[i] + 27.12032)/16.4)
    gyro_angle.append(gyro_tmp)
plt.scatter(time[0:100000],gyro_angle,label='gyro_angle drift',linewidth=1)
plt.show()
drift_rate = gyro_tmp / (dt*Gyro_x.shape[0])
print(drift_rate)
```
